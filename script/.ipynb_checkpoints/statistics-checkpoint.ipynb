```python
import os
import pickle
import joblib
from EntityPairItem import BertEntityPairItem
```

```python
path = "../data/"
with open(os.path.join(path, "train-token-term.jl"), "rb") as f:
    train = joblib.load(f)
print(f"train set triplet size: {len(train)}")
with open(os.path.join(path, "valid-token-term.jl"), "rb") as f:
    valid = joblib.load(f)
print(f"validation set triplet size: {len(valid)}")
with open(os.path.join(path, "EntityPairItems-disodiso-gold-500-token-term.jl"), "rb") as f:
    annotated = joblib.load(f)
print(f"annotated set triplet size: {len(annotated)}")
```

```python
# count sentences
count = {"train":{0:0, 1:0, 2:0, 3:0}, "valid":{0:0, 1:0, 2:0, 3:0} ,"annotated":{0:0, 1:0, 2:0, 3:0}}
for setname in count:
    for each in eval(setname):
        count[setname][int(each.label)] += len(each.sentences)
count
```

```python
# count triplet
count = {"train":{0:0, 1:0, 2:0, 3:0}, "valid":{0:0, 1:0, 2:0, 3:0} ,"annotated":{0:0, 1:0, 2:0, 3:0}}
for setname in count:
    for each in eval(setname):
        count[setname][int(each.label)] += 1
count
```

```python
# count entities
count = {"train":set(), "valid":set() ,"annotated":set()}
for setname in count:
    for each in eval(setname):
        count[setname].add(each.cui1)
        count[setname].add(each.cui2)
for key, item in count.items():
    print(key, len(item))
```

```python
with open(os.path.join(path, "graph.pkl"), "rb") as f:
    graph = pickle.load(f)
ent = set()
cnt = {0:0, 1:0, 2:0, 3:0}
for cui1, label, cui2 in graph:
    if label in cnt.keys():
        ent.add(cui1)
        ent.add(cui2)
        cnt[label] += 1
print(len(ent), cnt)
```
